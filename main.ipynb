```python
import os
import yaml
from pathlib import Path
from ultralytics import YOLO
import torch

dataset_root = Path("Vehicle_Detection_Image_Dataset")
train_images = list((dataset_root / "train" / "images").glob("*.jpg"))
val_images = list((dataset_root / "valid" / "images").glob("*.jpg"))

print(f"Training images: {len(train_images)}")
print(f"Validation images: {len(val_images)}")

corrected_data_yaml = {
    'train': os.path.abspath('Vehicle_Detection_Image_Dataset/train/images'),
    'val': os.path.abspath('Vehicle_Detection_Image_Dataset/valid/images'),
    'nc': 1,
    'names': ['Vehicle']
}

with open('data_corrected.yaml', 'w') as f:
    yaml.dump(corrected_data_yaml, f, default_flow_style=False)
```

```python
model = YOLO('models/yolov8n.pt')

training_args = {
    'data': 'data_corrected.yaml',
    'epochs': 10,
    'batch': 8,
    'imgsz': 320,
    'device': 'cpu',
    'project': 'vehicle_training',
    'name': 'vehicle_model',
    'cache': False,
    'workers': 4
}
```

```python
print("Starting training...")
results = model.train(**training_args)
print(f"Training complete. Model saved at: {results.save_dir}")
```

```python
from pathlib import Path
import os

trained_model_path = Path('vehicle_training/vehicle_model2/weights/best.pt')

if trained_model_path.exists():
    trained_model = YOLO(str(trained_model_path))
    cars_image = "cars.jpg"
    
    if os.path.exists(cars_image):
        results = trained_model.predict(
            source=cars_image,
            save=True,
            conf=0.25,
            project='runs',
            name='car_detection'
        )
        
        print(f"✅ Detection completed on {cars_image}")
        print(f"📁 Results saved in: runs/car_detection/")
        
        if results[0].boxes is not None:
            print(f"🔍 Detected {len(results[0].boxes)} vehicles:")
            for i, box in enumerate(results[0].boxes):
                conf = box.conf[0].item()
                print(f"   Vehicle {i+1}: {conf:.2f}")
        else:
            print("🔍 No vehicles detected")
    else:
        print(f"❌ Image '{cars_image}' not found")
else:
    print(f"❌ Model not found: {trained_model_path}")
```

```python
# Video detection using trained model
from pathlib import Path

trained_model_path = Path('vehicle_training/vehicle_model2/weights/best.pt')

if trained_model_path.exists():
    trained_model = YOLO(str(trained_model_path))
    video_file = "videoplayback.mp4"  # Change to your video file
    
    if os.path.exists(video_file):
        print(f"🎥 Processing video: {video_file}")
        
        results = trained_model.predict(
            source=video_file,
            save=True,
            conf=0.25,
            project='runs',
            name='video_detection',
            show_labels=True,
            show_conf=True
        )
        
        print("✅ Video processing completed!")
        print("📁 Results saved in: runs/video_detection/")
        print("🎬 Annotated video with detected vehicles saved")
        
    else:
        print(f"❌ Video '{video_file}' not found")
        print("📂 Available videos:", [f for f in os.listdir('.') if f.lower().endswith(('.mp4', '.avi', '.mov'))])
        
else:
    print(f"❌ Model not found: {trained_model_path}")
```

```python
# View .avi video with detected vehicles
import cv2
import os

video_path = "runs/video_detection/videoplayback.avi"

if os.path.exists(video_path):
    cap = cv2.VideoCapture(video_path)
    
    print(f"🎥 Playing: {video_path}")
    print("Press 'q' to quit")
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
            
        cv2.imshow('Vehicle Detection', frame)
        
        if cv2.waitKey(30) & 0xFF == ord('q'):
            break
    
    cap.release()
    cv2.destroyAllWindows()
    print("✅ Video closed")
else:
    print(f"❌ Video not found: {video_path}")
```
